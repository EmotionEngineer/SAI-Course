# 🌟 Building a Simple Chatbot in Python

Welcome to this interactive notebook! Today, we're creating a **basic chatbot** that can respond to greetings. This project will help you understand how simple AI models can interpret and respond to user input.

---

## 🤖 What is a Chatbot?

A **chatbot** is an AI-based tool that can simulate conversation with a user. Chatbots are used in customer service, online shopping, education, and more. Today, we’ll build a very simple chatbot that can respond to greetings like "hello" and "hi".

---

## 🛠️ Code Walkthrough

In the code below, we'll:
1. Take **user input** and convert it to lowercase for easy comparison.
2. Use `if` conditions to check for common greetings.
3. Return a **response** based on the input.

> **Tip:** Try to implement other responses!

---

```python
# Define a simple chatbot that responds to greetings

def simple_chatbot(user_input):
    user_input = user_input.lower()
    
    # Basic greetings
    if "hello" in user_input or "hi" in user_input:
        return "Hello! How can I help you today?"
    elif "how are you" in user_input:
        return "I'm just a bot, but I'm doing great! Thanks for asking."
    elif "bye" in user_input:
        return "Goodbye! Have a great day!"
    else:
        return "I'm sorry, I don't understand that. Can you try saying something else?"
```

```python
# Test the chatbot
user_message = input("You: ")
print("Bot:", simple_chatbot(user_message))
```
